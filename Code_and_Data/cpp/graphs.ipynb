```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import make_interp_spline
```

```python
gpu_split = [32, 64, 128, 256, 512, 1024]
gpu_50 = [0.14477, 0.153288, 0.140118, 0.176665, 0.297064, 0.567129]
gpu_1000 = [0.40615, 0.575967, 0.892954, 1.70297, 3.41548, 6.96135]

# less bad , no memory hits on total sums, no memory 
gpu_50_less_bad = [0.027359, 0.031498, 0.043755, 0.072804, 0.13499, 0.280885]
gpu_1000_less_bad = [0.172819, 0.17396, 0.265837, 0.536214, 1.13776, 2.44854]
```

```python
cpu_split = list(range(50, 1050, 50))
cpu_50 = [0.044785,0.076413,0.123216,0.172916,0.217207,0.276675,0.336116,0.385674,0.453788,0.542751,0.597307,0.681072,0.761365,0.866636,0.946079,1.04632,1.13809,1.25314,1.35938,1.44445]
cpu_1000 = [0.571198, 1.1521, 1.72845, 2.3312,2.88889,3.46318,4.22469,4.68135,5.26509,5.88943,6.51282,7.35058,7.9043,8.39347,9.07378,9.71356,10.5009,11.9117,12.0263,12.665]
```

```python

X_ = np.linspace(0, 1050, 500)

gpu_50_spline = make_interp_spline(gpu_split, gpu_50)
Y_gpu_50 = gpu_50_spline(X_)

gpu_50_lessbad_spline = make_interp_spline(gpu_split, gpu_50_less_bad)
Y_gpu_50_lessbad = gpu_50_lessbad_spline(X_)

gpu_1000_spline = make_interp_spline(gpu_split, gpu_1000)
Y_gpu_1000 = gpu_1000_spline(X_)

gpu_1000_lessbad_spline = make_interp_spline(gpu_split, gpu_1000_less_bad)
Y_gpu_1000_lessbad = gpu_1000_lessbad_spline(X_)

cpu_50_spline = make_interp_spline(cpu_split, cpu_50)
Y_cpu_50 = cpu_50_spline(X_)

cpy_1000_spline = make_interp_spline(cpu_split, cpu_1000)
Y_cpu_1000 = cpy_1000_spline(X_)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))

fig.suptitle('CPU & GPU Runtimes on 01 Knapsack')
ax1.set_title("50 item Knapsack")
ax1.set_ylabel("Time (s)")
ax1.set_xlabel("Population Size")
ax1.plot(X_, Y_gpu_50, label='1st GPU')
ax1.plot(X_, Y_cpu_50, label='CPU')
ax1.plot(X_, Y_gpu_50_lessbad, label='2nd GPU')
ax1.legend(loc='upper left')

ax2.set_title("1000 item Knapsack")
ax2.set_ylabel("Time (s)")
ax2.set_xlabel("Population Size")
ax2.plot(X_, Y_gpu_1000, label='1st GPU')
ax2.plot(X_, Y_cpu_1000, label='CPU')
ax2.plot(X_, Y_gpu_1000_lessbad, label='2nd GPU')
ax2.legend(loc='upper left')

fig.tight_layout()
plt.show()
```
